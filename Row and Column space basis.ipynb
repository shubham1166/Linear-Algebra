```python
import numpy as np
np.set_printoptions(precision=16)
import math
```

```python
#algorithm to find the index of first non-zero element in a column from position start_row 
def find_nonzero_element_index(col, start_row):
    #print(col)
    ret_index = -1
    for k in np.arange(start_row,col.size):
        #print('is close to zero: %s' %(math.isclose(np.abs(col[k]), 0., abs_tol=1e-13)))
        if not math.isclose(np.abs(col[k]), 0., abs_tol=1e-13):
            ret_index = k
            break
    #print ('ret index: %d' %(ret_index))
    return ret_index
```

```python
#algorithm to convert a matrix A to rref
#algorithm to convert a matrix A to rref
def find_rref(A):
    #print(A.shape)
    m = (A.shape)[0] #num rows
    n = (A.shape)[1] #num cols
    print('num rows: %d' %(m))
    print('num cols: %d' %(n))

    j = 0
    r = 0
    
    B=A.copy()
    
    for j in np.arange(0,n):
        non_zero_row_index = find_nonzero_element_index(B[:,j],r)
        #print('r: %d' %(r))
        if non_zero_row_index != -1: 
            #write the code to make the pivot element in the non-zero row as 1 
            #and to make all other elements in the pivot column as zero  
            
            #If non_zero index is greater than the r then we will interschange the two rows
            if non_zero_row_index>r:
                c=r
                B[[non_zero_row_index,r]]=B[[r,non_zero_row_index]]
            else:
                c=non_zero_row_index
            B[c,:]=B[c,:]/B[c,j]
            
            
            #performimg the row operations
            for i in range(m):
                if i!=c:
                    B[i,:]=B[i,:]-B[i,j]*B[c,:]
            
            #you can check your code by printing the intermediate matrix using the following statement
            #print('rref process in col: %d ' %(j))
            #print(A)
            r = r+1
    return (B)

```

```python
#returns the non-zero row indices and dimension of rowspace(A)
def find_rowspace_basis_dim(rref_A):
    m = (rref_A.shape)[0]
    non_zero_indices = [] 
    start_index = 0
    for i in np.arange(0,m):
        non_zero_pos = find_nonzero_element_index(rref_A[i].T,start_index)
        if non_zero_pos != -1:
            start_index = non_zero_pos
            non_zero_indices.append(non_zero_pos)
        else:
            break
    return non_zero_indices, len(non_zero_indices)
```

# In the code for finding the basis and dimension of the column space, we'll use the row reduced echlon form of the given matrix and will extract the pivot columns that will give us the basis and the number of pivot columns will give us the dimension of the basis of the column space

```python
# defining a function to extract set of linearly independent columns/basis of column space with the help of RREF

def extract_column_basis(A):
    #B=find_rref(A)
    #Finding the pivot columns index
    B=find_rref(A)
    #print('The rref of the given matrix is : ',)
    x=[]
    (m,n)=B.shape
    for i in range(m):
        for j in range(n):
            if B[i,j]==1:
                x.append(j)
                break    
    
    column_basis=[]
    for k in x:
        column_basis.append(A[:,k])
    
    return column_basis, len(column_basis)
```
