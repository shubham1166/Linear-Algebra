```python
import numpy as np
np.set_printoptions(precision=16)
import math
```

```python
#algorithm to find the index of first non-zero element in a column from position start_row 
def find_nonzero_element_index(col, start_row):
    #print(col)
    ret_index = -1
    for k in np.arange(start_row,col.size):
        #print('is close to zero: %s' %(math.isclose(np.abs(col[k]), 0., abs_tol=1e-13)))
        if not math.isclose(np.abs(col[k]), 0., abs_tol=1e-13):
            ret_index = k
            break
    #print ('ret index: %d' %(ret_index))
    return ret_index
```

```python
#algorithm to convert a matrix A to rref
#algorithm to convert a matrix A to rref
def find_rref(A):
    #print(A.shape)
    m = (A.shape)[0] #num rows
    n = (A.shape)[1] #num cols
    print('num rows: %d' %(m))
    print('num cols: %d' %(n))

    j = 0
    r = 0
    
    B=A.copy()
    
    for j in np.arange(0,n):
        non_zero_row_index = find_nonzero_element_index(B[:,j],r)
        #print('r: %d' %(r))
        if non_zero_row_index != -1: 
            #write the code to make the pivot element in the non-zero row as 1 
            #and to make all other elements in the pivot column as zero  
            
            #If non_zero index is greater than the r then we will interschange the two rows
            if non_zero_row_index>r:
                c=r
                B[[non_zero_row_index,r]]=B[[r,non_zero_row_index]]
            else:
                c=non_zero_row_index
            B[c,:]=B[c,:]/B[c,j]
            
            
            #performimg the row operations
            for i in range(m):
                if i!=c:
                    B[i,:]=B[i,:]-B[i,j]*B[c,:]
            
            #you can check your code by printing the intermediate matrix using the following statement
            #print('rref process in col: %d ' %(j))
            #print(A)
            r = r+1
    return (B)

```

## The function extract_pivot_column is a function that is made by making some changes in the function to find the column space from the last assignment

```python
#Defining a function to find the pivot columns in the matrices , so we know that only those columns are the one which are pivot columns
def extract_pivot_column(A):
    #B=find_rref(A)
    #Finding the pivot columns index
    B=find_rref(A)
    #print('The rref of the given matrix is : ',)
    x=[]
    (m,n)=B.shape
    for i in range(m):
        for j in range(n):
            if B[i,j]==1:
                x.append(j)
                break    
    
    #column_basis=[]
    #for k in x:
    #    column_basis.append(A[:,k])
    
    return x
```

## Now Let us define the function to find the null space of a matrix , it is using the above functions to find the pivot columns and the rref of the matrix. 

```python
# Now lets find the Null space of this matrix

def null_space(B):
    #Finding the rref of the matrix B
    A=find_rref(B)
    #b is an array that will give us the index of the pivot columns
    b=extract_pivot_column(A)
    #counter=0
    null_space=[]
    m,n=A.shape
    #We know that the dim of the null space is n-dim of the column space or n- (no of pivot columns) ,  so we are running loop for n-len(b) times basically , so that every time it gives a vector in the null space
    for i in range(n):
        if i not in b:
            #Firsty initializing a matrix [0] of n*1
            dummy=np.zeros([n,1])
            #we can take any non-zero value to dummy[i]
            #For the sake of convinience, giving it a value of 1
            #Basically , we know that ith index variable is independent variable , so we are giving it 1 value
            #Now in the next loop, we will find all the dependent variable that are dependent on xi or this ith variable
            dummy[i]=1
            dummy2=dummy.copy()
            for j in b:
                k=0
                #Finding the row in which the jth pivot column exists
                for t in range(m):
                    if A[t,j]==1:
                        k=t
                        #Giving the j-th variable this value as it is dependent on the i th variable and the ith variable is given thevalue of 1
                        dummy2[j]= -A[k,i]
                        break
            null_space.append(dummy2)
            #The Function is returning the basis of null space(an array of Rn*1 vectors basically) and the dimension of the null space
    
    return null_space,len(null_space)
    
```
